#**Introdução ao processamento de imagens**

*   Abrindo e exibindo uma imagem


```
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```
image = cv2.imread("/content/imagem1.jpg")

print(image.shape)
plt.imshow(image)
plt.show()
```

```
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Converte uma imagem de um espaço de cores para outro
plt.imshow(gray, cmap="gray")
plt.show()
```
