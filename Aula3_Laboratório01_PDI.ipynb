# **Transformações de Imagens**

```python
# importando as bilbiotecas
import cv2 as cv
import matplotlib.pylab as plt
from google.colab.patches import cv2_imshow
```

```python
# exibindo a imagem
imagem = cv.imread('/content/ponte.jpg')
cv2_imshow(imagem)
```

* Criando uma nova imagem a partir de um pedaço da imagem original

```python
recorte = imagem[100:200, 100:200] # linha 101 até a linha 200 na coluna 101 até a coluna 200
plt.imshow(recorte)
```

* Redimensionamento -> Resize

```python
import imutils
import numpy as np

# No canto inferior esquerdo da imagem é possível notar a imagem redimensionada
proporcao = 100.0 / imagem.shape[1]
tamanho_novo = (100, int(imagem.shape[0] * proporcao))
img_redimensionada = cv.resize(imagem, tamanho_novo, interpolation = cv.INTER_AREA)
cv2_imshow(img_redimensionada)
```

```python
# outra forma de redimensionar a imagem
img_redimensionada1 = imagem[::2,::2]
cv2_imshow(img_redimensionada1)
```

* Espelhando uma imagem -> flip

```python
flip_horizontal = imagem[::-1,:] 
cv2_imshow(flip_horizontal)
```

```python
flip_vertical = imagem[:,::-1]
cv2_imshow(flip_vertical)
```

```python
flip_hv = imagem[::-1,::-1]
cv2_imshow(flip_hv)
```

* Rotacionando uma imagem

```python
(altura, largura) = imagem.shape[:2] # encontra a altura e a largura
centro = (largura // 2, altura // 2) # encontra o centro

M = cv.getRotationMatrix2D(centro, 30, 1.0) # 30 graus
img_rotacionada = cv.warpAffine(imagem, M, (largura, altura))

cv2_imshow(img_rotacionada)
```

* Máscaras

```python
mascara = np.zeros(imagem.shape[:2], dtype = "uint8")

(cX, cY) = (imagem.shape[1] // 2, imagem.shape[0] // 2)

cv.circle(mascara, (cX, cY), 100, 255, -1)

img_com_mascara = cv.bitwise_and(imagem, imagem, mask = mascara)

cv2_imshow(img_com_mascara)
```

* Exibindo canais RGB

```python
(canalAzul, canalVerde, canalVermelho) = cv.split(imagem)

zeros = np.zeros(imagem.shape[:2], dtype = "uint8")
```

```python
cv2_imshow(cv.merge([zeros, zeros, canalVermelho]))
```

```python
cv2_imshow(cv.merge([zeros, canalVerde, zeros]))
```

```python
cv2_imshow(cv.merge([canalAzul, zeros, zeros]))
```
