```python
# a)
import cv2 as cv 
from google.colab.patches import cv2_imshow 
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
from skimage.transform import resize, rotate 
```

```python
# b)
imagem = cv.imread("/content/exercicio01.jpg") #Abre uma imagem com o OpenCV
cv2_imshow(imagem) #Exibe a imagem
```

```python
# c) 
print(imagem)
```

```python
# d)
print(imagem[0,0])
```

```python
# e)
print("Tamanho da imagem: ", imagem.shape) #Exibe altura, largura e canais
```

```python
# f)
from skimage.color import rgb2gray, rgb2gray # calcula a luminância de uma imagem RGB

#Converter imagem para o espaço de cor grayscale
img_gray = rgb2gray(imagem)

plt.imshow(img_gray, cmap = 'gray')
plt.show()
```

```python
# g) 
print(img_gray.shape)
```

```python
#h)
print(imagem.max())
print(imagem.min())
print(imagem.mean())

print(img_gray.max())
print(img_gray.min())
print(img_gray.mean())
```

```python
#i)

a,l,c = imagem.shape
imagem_red = resize(imagem, (int(a/5), int(l/5)))
plt.imshow(imagem_red)
print(imagem.shape, " Imagem normal ")
print(imagem_red.shape , " Imagem redimensionada ")
```

```python
#j)

imagem_rotacionada = rotate(imagem, 90, resize = True)
plt.imshow(imagem_rotacionada)
```

```python
imagem_rotacionada = rotate(imagem, -90, resize = True)
plt.imshow(imagem_rotacionada)
```

```python
imagem_rotacionada = rotate(imagem, -180, resize = True)
plt.imshow(imagem_rotacionada)
```
