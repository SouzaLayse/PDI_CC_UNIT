# **FUNÇÕES**

* Pequenos trechos de código reutilizáveis
* Sequência de instruções que computa um ou mais resultados (parâmetros)
* Permitem dar um nome a um bloco de comandos e executar esse bloco, a partir de qualquer lugar do programa

A sintaxe é:

```
def <nome da função> (<parâmetros>)
    <comandos da função>
    return <resultado>
```


**Exemplo:** Criar uma função chamada olá. Essa função deve imprimir a frase olá mundo

````python
def ola():
  print('Olá mundo!!!')
````

````python
ola()
````

* ***Parâmetros (opcional):*** variáveis que podem ser incluídas nos parênteses das funções.
* ***Argumentos:*** valores passados para as variáveis.


**Exemplo:** Criar uma função para imprimir o menor valor entre 2 números (a,b)

````python
def menor(a,b):
  if a < b:
    print(a)
  else:
    print(b)
````

````python
# menor(a,b)
menor(15,17)
````

````python
# menor(a,b)
menor(17,15)
````

* Toda variável utilizada em uma função não será acessível por outras funções ou até mesmo pelo programa principal (shell). 

* Se tiver uma variável com o mesmo nome fora da função, essa variável é independente, ou seja, a variável da função é uma e a variável fora da função é outra

**Exemplo:** Função para realizar a soma de dois números (a, b)

````python
def soma(a,b):
  total = a + b
  print('Total da soma da função é: ', total)
````

````python
total = 8
soma(17,29)
print('Total fora da função é: ', total)
````

* Para compartilhar uma variável em diversas funções e no programa principal (shell), a variável deve ser definida como global

**Exemplo:** Função para realizar a soma de dois números (a, b)

````python
def soma(a,b):
  global total
  total = a + b
  print('Total da soma da função é: ', total)
````

````python
global total

total = 41
soma(7,5)
print('Total fora da função é: ', total)
````

* Retornar valores e encerrar uma função -> return

*Obs.: Se não for declarado um valor de retono, a função retorna o valor None*

**Exemplo:** Criar três funções (subtração, multiplicação e divisão) e retornar o resultado dessas funções

````python
def subtracao(a,b):
  resultado = a - b
  return resultado

def multiplicacao(a,b):
  resultado = a * b
  return resultado

def divisao(a,b):
  resultado = a / b
  return resultado
````

````python
subtracao(11,5)
````

````python
multiplicacao(29,6)
````

````python
divisao(45,5)
````

**Valor padrão**

* Para definir um valor padrão nos parâmetros de uma função, é necessário omitir o valor na chamada da função, dessa forma, a variável assume o valor padrão

**Exemplo:** Calcular juros com a taxa de 15%

````python
def calculaJuros(valor, taxa=15):
  juros = valor*taxa/100
  return juros
````

````python
calculaJuros(1900)
````

**Múltiplos argumentos** 

*args
* Permite passar um número indeterminado de parâmetros para uma função
* Os argumentos são passados por posição e sem uma chave para identificar


**Exemplo:** Somar números

````python
def soma(*args):
   return sum(args)
````

**kwargs
* Permite passar um número indeterminado de parâmetros para uma função
* Os argumentos são passadas chaves, como fazemos nos dicionários


**Exemplo:** Estados das regiões do Brasil

````python
def estadosRegioes(**kwargs):
  for regiao, estado in kwargs.items():
    print('Na região', regiao, 'tem o estado', estado)
````

````python
estadosRegioes(sul ='Paraná', norte ='Acre', nordeste ='Sergipe', sudeste = 'São Paulo', centroOeste = 'Goiás')
````

# **EXCECÕES**

*   São trechos de códigos responsáveis por tomar atitudes e dar respostas quando ocorrem  exceções
*   Para capturar uma exceção utilizamos 

```
try
  [código]
except [exceções]:
  [código que será executado no caso da exceção]
```

**Exemplo:** Calcular a raiz quadrada de um número (positivo e negativo)

````python
# importando o módulo math 
import math 

# criando um variável para receber o número 
numero = 25
#numero = -25

# tratando as exceções com os blocos try e except 
try:
  # criando uma variável que vai receber a raiz quadrada do número
  result = math.sqrt(numero)
  # imprimindo o número e sua raiz quadrada
  print('O número é ', numero, 'e a raiz quadrada desse numéro é ', result)
except Exception:
  # imprimindo a mensagem de exceção
  print ('O número é negativo, logo não é possível calcular a raiz quadrada!')
````
