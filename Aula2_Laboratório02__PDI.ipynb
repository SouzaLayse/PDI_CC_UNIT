* Como desenhar usando o OpenCV

```python
# importando as bibliotecas 
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#Criar matriz, preenchido com zero, de 500 X 500 com três canais
tela = np.zeros((500, 500, 3), dtype="uint8") # integer (0 to 255)
plt.imshow(tela)
plt.show()
```

```python
# Criando cores
vermelho = (255, 0, 0)
verde = (0, 255, 0)
azul = (0, 0, 255)

# Criando linhas na tela -> cv2.line(imagem, inicio, fim, cor, largura)
cv2.line(tela, (0,0), (500, 500), verde)
plt.imshow(tela)
plt.show()
```

```python
cv2.line(tela, (0,500), (500, 0), vermelho)
plt.imshow(tela)
plt.show()
```

```python
cv2.line(tela, (0,250), (500, 250), (128, 50, 130), 3) # linha roxa
plt.imshow(tela)
plt.show
```

```python
#Desenhando retangulos
cv2.rectangle(tela, (100,100), (200, 400), azul, 5)
plt.imshow(tela)
plt.show()
```

```python
cv2.rectangle(tela, (100,200), (200, 300), azul, -1)
plt.imshow(tela)
plt.show()
```

```python
cv2.rectangle(tela, (200,100), (300, 400), (0,200, 190), -1)
plt.imshow(tela)
plt.show()
```

```python
#Desenhando circulos
cv2.circle(tela, (400, 250), 60, (255, 255, 255), 3)
plt.imshow(tela)
plt.show()
```

```python
cv2.circle(tela, (400, 250), 60, verde, -1)
plt.imshow(tela)
plt.show()
```

```python
#Circulos dentro de circulos
telanova = np.zeros((400, 400, 3), dtype="uint8") # integer (0 to 255)

(centroX, centroY) = (telanova.shape[1]//2, telanova.shape[0]//2)

for r in range(0, 175, 25):
  cv2.circle(telanova, (centroX, centroY), r, (255, 255, 255), 3)

plt.imshow(telanova)
plt.show()
```

* Operações Bitwise

AND

OR

XOR

NOT

```python
telaretangulo = np.zeros((400, 400), dtype="uint8") # integer (0 to 255)

plt.imshow(telaretangulo)
plt.show()
```

```python
cv2.rectangle(telaretangulo, (25, 25), (275, 275), 255, -1)

plt.imshow(telaretangulo)
plt.show()
```

```python
telacirculo = np.zeros((400,400), dtype="uint8") # integer (0 to 255)
cv2.circle(telacirculo, (150, 150), 150, 255, -1)
plt.imshow(telacirculo)
plt.show()
```

```python
# Operador AND
bitwiseAnd = cv2.bitwise_and(telaretangulo, telacirculo)
plt.imshow(bitwiseAnd)
plt.show()
```

```python
# Operador OR
bitwiseOr = cv2.bitwise_or(telaretangulo, telacirculo)
plt.imshow(bitwiseOr)
plt.show()
```

```python
# Operador XOR
bitwiseXor = cv2.bitwise_xor(telaretangulo, telacirculo)
plt.imshow(bitwiseXor)
plt.show()
```

```python
# Operador NOT
bitwiseNot = cv2.bitwise_not(telaretangulo, telacirculo)
plt.imshow(bitwiseNot)
plt.show()
```
